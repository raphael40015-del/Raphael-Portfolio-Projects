```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r"C:\Users\rapha\Documents\1991-2020 Temp and Precipation.xlsx")
df.head(12)
```

```python
# Data Frame
# Select only the columns you want
df = df[['Month',
        'Temp _Avg',
        'Temp_ Std',
        'Temp_Max',
        'Temp_Min',
        'Rainfall',
        'Snowfall',
        'Precipitation']]

# Now check first 5 rows
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))

ax.set_xlabel("Months", fontsize=12)
ax.set_ylabel("Temperature(Celsius)", fontsize=12)
ax.set_title("1991-2020 Monthly Temperature Statistics", fontsize=15)

ax.plot(df['Month'], df['Temp _Avg'], linestyle='--', marker='o', color='green', label="Temperature Average")
ax.plot(df['Month'], df['Temp_ Std'], linestyle='--', marker='o', color='red', label="Temperature Standard Deviation" )
ax.plot(df['Month'], df['Temp_Max'], linestyle='--', marker='o', color='blue', label="Temperature Max" )
ax.plot(df['Month'], df['Temp_Min'], linestyle='--', marker='o', color='orange', label="Temperature Min")

ax.legend()
ax.grid()


```

```python
import numpy as np

# Set positions
x = np.arange(len(df['Month']))  # positions for months
width = 0.35  # bar width

plt.figure(figsize=(12,6))
plt.bar(x - width/2, df['Rainfall'], width, label='Rainfall (mm)', color='blue')
plt.bar(x + width/2, df['Snowfall'], width, label='Snowfall (cm)', color='lightgray')

plt.title("Monthly Rainfall vs Snowfall (1991-2020)")
plt.xlabel("Month")
plt.ylabel("Precipitation")
plt.xticks(x, df['Month'], rotation=45)
plt.legend()
plt.show()
```
